### Importing the modules 

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

### Reading the Data

```python
corona_data= pd.read_csv("Datasets/covid19_Confirmed_dataset.csv")
corona_data.head(10)
```

```python
corona_data.shape
```

```python
corona_data.drop(["Lat","Long"], axis=1, inplace=True)
```

```python
corona_data.head()
```

```python
corona_data_aggregated=corona_data.groupby("Country/Region").sum()
```

```python
corona_data_aggregated.shape
```

### Visualizing data related to a country

### Finding the maxmimum infection rate

```python
corona_data_aggregated.loc["China"].diff().max()
```

```python
corona_data_aggregated.loc["Italy"].diff().max()
```

```python
corona_data_aggregated.loc["Spain"].diff().max()
```

```python
countries=list(corona_data_aggregated.index)
max_infection_rates=[]
for c in countries :
    max_infection_rates.append(corona_data_aggregated.loc[c].diff().max())
corona_data_aggregated["max_infection_rate"]=max_infection_rates
```

```python
corona_data_aggregated.head()
```

### A new dataframe with only needed column 

```python
corona_data_v2= pd.DataFrame(corona_data_aggregated["max_infection_rate"])
```

```python
corona_data_v2.head()
```
